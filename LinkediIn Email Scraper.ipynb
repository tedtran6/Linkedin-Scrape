```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
#explicit wait stuff
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import csv
```

```python
class LinkedInBot:
    #init class
    def __init__(self, username, password):
        option = webdriver.ChromeOptions()
        #option.add_argument('headless')
        
        self.username = username;
        self.password = password;

        self.driver = webdriver.Chrome(r"C:\Users\tedtr\Anaconda Projects\chromedriver.exe", options=option)
        
    #close the browser
    def closebrowser(self):
        self.driver.close()
        
    def waiting(self, xpath):
        try:
            driver.find_element_by_xpath(xpath)
        except:
            time.sleep(1)
            waiting(xpath)
        
    #login function
    def login(self):
        driver = self.driver
        driver.get("https://www.linkedin.com/login/")
        #driver.get("https://www.instagram.com/")
        
        #Username Input
        username_elem = WebDriverWait(driver, 10).until(EC.presence_of_element_located(('id', 'username')))
        #username_elem = driver.find_element_by_xpath("//*[@id='username']")
        username_elem.click()
        username_elem.send_keys(self.username)
        
        #Password input
        password_elem = driver.find_element_by_xpath("//*[@id='password']")
        password_elem.click()
        password_elem.send_keys(self.password)
        
        login_elem = driver.find_element_by_xpath("//*[@id='app__container']/main/div/form/div[3]/button")
        login_elem.click()
        
    #variables:
    # pages_to_search: how many pages you go through during search
    # title: Title of the persons Job
    def searchSpecifics(self, pages_to_search, title, csv_name):
        
        
        driver = self.driver
        driver.get("https://www.linkedin.com/search/results/people/?origin=DISCOVER_FROM_SEARCH_HOME")
        
        time.sleep(2)
        
        # get rid of messaging section
        messaging = driver.find_element_by_xpath("//button[@class='msg-overlay-bubble-header__button truncate ml2']")
        messaging.click()
        
        all_filters = driver.find_element_by_xpath("/html/body/div[5]/div[4]/div[3]/div/div[1]/header/div/div/div[2]/button/span");
        
        all_filters.click()
        
        time.sleep(2)
        
        dallas_elem = driver.find_element_by_xpath("//*[@id='locations-facet-values']/fieldset/ol/li[3]/label")
        dallas_elem.click()
        
        title_elem = driver.find_element_by_xpath("//*[@id='search-advanced-title']")
        title_elem.click()
        title_elem.send_keys(title)
        
        #use this as an example of how to properly select an element with selenium
        apply_elem = driver.find_element_by_xpath("//button[@data-control-name='all_filters_apply']")
        apply_elem.click()
        
        time.sleep(3)
        
        # go through pages
        for i in range(pages_to_search):
        
            #scroll down
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

            time.sleep(2)
            
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            
            time.sleep(2)

            #get the list of 10 people on the page
            people = driver.find_elements_by_xpath("//a[@data-control-name='search_srp_result'][@class='search-result__result-link ember-view']")
            condensedPeople = []

            #people get listed twice, once for their picture and once for their name.
            for i in range(len(people)):
                if i%2 == 1:
                    condensedPeople.append(people[i])

            for i in condensedPeople:
                print(i.get_attribute("href"))
                
                user_open_script = "window.open('" + str(i.get_attribute("href")) + "', 'new window')"
                driver.execute_script(user_open_script)
                driver.switch_to_window(driver.window_handles[1])
                
                time.sleep(1)
                
                # Name of Person
                name = driver.find_element_by_xpath("//li[@class='inline t-24 t-black t-normal break-words']")
                name_text = name.text
                name_parts = name_text.split()
                first_name = name_parts[0]
                last_part = ''
                
                for i in name_parts[1:]:
                    last_part += str(i) + ' '
                    
                print(first_name)
                print(last_part)
                
                # Contact Info Button
                contact_info = driver.find_element_by_xpath("//a[@data-control-name='contact_see_more']")
                contact_info.click()
                
                time.sleep(1)
                
                email_is_present = len(driver.find_elements_by_xpath("//section[@class='pv-contact-info__contact-type ci-email']")) > 0
                
                if(email_is_present):
                    email_section = driver.find_element_by_xpath("//section[@class='pv-contact-info__contact-type ci-email']")
                    email = email_section.find_element_by_tag_name('a')
                    email_text = email.get_attribute('href')
                    split_email = email_text.split(':')
                    real_email = split_email[1]
                    print(real_email)
                    print('email found')
                    # start the CSV
                    with open('persons.csv', 'w') as csvfile:
                        filewriter = csv.writer(csvfile, delimiter=',', quotechar='|', quoting=csv.QUOTE_MINIMAL)
                        
                        filewriter.writerow([first_name, last_part, real_email])
                try:
                    print('hello')
                except:
                    print('no email')
                
                driver.close()
                driver.switch_to_window(driver.window_handles[0])
                
            
            
            # next page
            # all I needed to do is wait a little. The NEXT button wasn't showing up on the screen yet... I need to figure out the proper wait
            next_elem = driver.find_element_by_xpath("//button[@aria-label='Next']")
            next_elem.click()
            
    def scrape_info():
        return ''
```
